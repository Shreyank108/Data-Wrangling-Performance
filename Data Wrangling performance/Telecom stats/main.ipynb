```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("borad new.csv")
```

```python
data.columns = data.columns.str.strip()
```

```python
print(data.isnull().sum())
```

```python
data['TIME'] = pd.to_datetime(data['TIME'], format='%Y')
```

```python
plt.figure(figsize=(12, 8))
```

```python
for country in data['Country'].unique():
    country_data = data[data['Country'] == country]
    plt.plot(country_data['TIME'], country_data['Value'], label=country)

plt.title('Mobile Data Usage per Mobile Broadband Subscription Over Time')
plt.xlabel('Year')
plt.ylabel('Mobile Data Usage (GB per month)')
plt.legend()
plt.grid(True)
plt.show()
```
